```python
data.Geography.replace(['France','Germany','Spain'],[0,1,2],inplace = True)
data.Gender.replace(['Male','Female'],[0,1],inplace =True)
```

```python
data.head()
```

## Extract Feather and Label

```python
x = data[['CreditScore','Geography','Gender','Age',	'Tenure',	'Balance',	'NumOfProducts',	'HasCrCard',	'IsActiveMember',	'EstimatedSalary']]
y = data[['Exited']]
```

```python
x.shape
```

```python
y.shape
```

## Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,shuffle=True)
```

## Import Sequential and Dense

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
xtrain.shape
```

```python
model.add(Dense(units = 8, activation = 'relu', input_dim = 10 ))
model.add(Dense(units = 6, activation = 'relu' ))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
history = model.fit(xtrain, ytrain, validation_split=0.2, epochs=100, verbose=1)
```

## Evaluation

```python
model.evaluate(xtest,ytest)
predictions = model.predict(xtest)
predictions
```
